```
import numpy as np

# atributos:
# 1. amosta [[x1,x2,x3,...,xn]]; 2. respostas [r1,r2,r3,...,rn]; 3. learning rate; iterações (maxima)

# metodos:
# 1. treino; 2. teste

class Perceptron:
  
  def __init__(self, amostra, resultados, taxa_aprendizado, iteracoes):
    self.amostra = amostra
    self.resultado = resultados
    self.taxa_aprendizado = taxa_aprendizado
    self.iteracoes = iteracoes
    
    self.peso = []
    self.num_parametros = self.amostra.shape[1]
    
    self.amostra_treino = self.amostra#self.amostra[:self.amostra.shape[0]//2,:]
    self.amostra_teste = self.amostra#self.amostra[self.amostra.shape[0]//2:,:]

    self.resultado_treino = self.resultado#self.resultado[:self.resultado.shape[0]//2]
    self.resultado_teste = self.resultado#self.resultado[self.resultado.shape[0]//2:]

    
    
    
  def treino(self):
    
    # vetor de pesos iniciais (aleatórios)
    for i in range(self.num_parametros):
      self.peso.append(np.random.random_sample())
    print('- Pesos iniciais:', self.peso)
    
    # declarando contador de iteracoes
    it = 0
    
    # iniciando treino
    print('TREINO --------------------------------------------------------------')
    while True:
      error = False
      
      print('#Iteracao numero', it)
      for i in range(self.amostra_treino.shape[0]):

        # calculo de output
        saida = 0 
        for j in range(len(self.peso)):
          saida += self.peso[j] * self.amostra_treino[i][j]
        print('- Pesos:',self.peso)
        print('- Saida:', saida)
        
        # avaliando com a função de perda
        # f(x) = <( 1 se x >= 0
        #         ( 0 se x < 0
        previsao = 1 if saida >= 0 else 0
        print('- Previsão, Resposta:', previsao, self.resultado[i])

        # testando se havera necessidade de atualizar pesos
        if previsao == self.resultado[i]:
          #acertou
          print('ACERTOU!!')
          continue
        else:
          #errou
          print('ERROU!!')
          # calculando erro
          erro = self.resultado[i] - previsao

          # atualizando pesos
          for j in range(len(self.peso)):
            self.peso[j] = self.peso[j] + erro * self.amostra_treino[i][j] * self.taxa_aprendizado
          # infere que ainda há erro
          error = True
      
      print('\n')
      
      # contador de iteracoes
      it += 1
      if error == False or it == self.iteracoes:
        print('#numero de iterações',it)
        break

      
      
  def teste(self, tipo1, tipo2):
    
    # iniciando teste
    print('TESTE ---------------------------------------------------------------')
    for i in range(self.amostra_teste.shape[0]):
      
      # calculo de output
      saida = 0 
      for j in range(len(self.peso)):
        saida += self.peso[j] * self.amostra_teste[i][j]
      

      # avaliando com a função de perda
      # f(x) = <( 1 se x >= 0
      #         ( 0 se x < 0
      previsao = 1 if saida >= 0 else 0
      
      print('A amostra pertence ao grupo:',tipo1 if previsao == 1 else tipo2,'  ACERTOU' if previsao == self.resultado[i] else '  ERROU','   - Gabarito:',self.resultado[i])
    
      

     

#REFERENCIA:
#https://juliocprocha.blog/2017/07/27/perceptron-para-classificacao-passo-a-passo/comment-page-1/?unapproved=509&moderation-hash=87530cc77503403f8700aad34b1f76bf#comment-509
#https://juliocprocha.blog/2017/08/14/perceptron-em-python/
#https://pt.wikipedia.org/wiki/Perceptron
# https://www.youtube.com/watch?v=9nWMXIXNGdU 22min
```

```
# DADOS BINARIOS

import pandas as pd
import numpy as np

dataSet = pd.read_csv('iris_2_resumida.csv')

# organizando dados de saida
labels = np.array(list(set(dataSet.species)))
labels = np.flip(labels, 0)
saidas_bi = []
for x in range(len(dataSet.species)):#O for começa de 1 pois na primeira linha esta o cabeçario.
             
            for y in range(len(labels)):#dou um for na variavel labels
                if labels[y] in dataSet.species[x]:#avalio qual classe esta contida na linha
                    saidas_bi.append(y) #Substituo a string por um float no caso 0 ou 1.

# organizando amostra
dataSet = dataSet.drop(columns = ['species','petal_length', 'petal_width'])
amostra_bi = np.array(dataSet)

# cria uma rede Perceptron
rede = Perceptron(amostra_bi, saidas_bi,	0.1,100)


# treina a rede
rede.treino()

# testando a rede

rede.teste( 'A', 'B')
```
